## Ordered dithering

```python
def ordered_dithering(image,pattern_law, path):
    """
    Arguments
        image: a 2D numpy array
        pattern_law: a 2D matrix containing the values to generate the
            patterns.
        path: a string, if 'a' or 'b' it will use the pattern a as
        described in the assigment.
        
    """
    # Create the patterns based on its law
    num_patterns = pattern_law.shape[0]*pattern_law.shape[1] + 1
    patterns = np.array([np.where(pattern_law < i, 1, 0) for i in range(num_patterns)])
    results = np.zeros((pattern_law.shape[0]*image.shape[0], pattern_law.shape[1]*image.shape[1]))
    
    # Normalize the image to range(0,num_patterns -1)
    image = image/max(image.max(), 1) * (num_patterns-1)
    image = image.astype(np.int32)
    
    # Choose the path
    num_rows, num_columns = image.shape
    if path == 'a':
        intervals = [range(num_columns), range(num_columns)]
        patterns = [patterns,patterns]
    if path == 'b':
        intervals = [range(num_columns), range(num_columns,0)]
        inverted_patterns = np.array([np.flip(pattern) for pattern in patterns])
        patterns = [patterns,inverted_patterns]
    # Run
    for i in range(num_rows):
        interval = intervals[i%2]
        line = []
        for j in range(num_columns):
            pattern = patterns[i%2][image[i,j]]
            results[pattern_law.shape[0]*i:pattern_law.shape[0]*i + pattern_law.shape[0],\
               pattern_law.shape[1]*j:pattern_law.shape[1]*j + pattern_law.shape[1]] = pattern    
    return results
```

```python
def floyd_steinberg_dithering(image, path):
    """Apply Floyd Steinsberg algorithm using the path described in
    figure(a).
    Arguments:
        image: a gray gray scaled single channel image
        path: a string specifying one of the next paths:
            'a':
            'b1': uses the path b
            'b2': uses the path b and reflect the pattern
    Return:
        image: the image after aplied the algorithm
    """
    num_lines, num_columns = image.shape
    if path == 'a':
        intervals = [range(2, num_columns-2), range(num_columns-2)]
        flip = False
    if path == 'b1':
        intervals = [range(2,num_columns-2), range(num_columns-2,2)]
        flip = False
    if path == 'b2':
        intervals = [range(2,num_columns-2), range(2,num_columns-2)]
        flip = True
    for i in range(2,num_lines-2):
        interval = intervals[i%2] 
        for j in interval:
            cut_image = round(image[i,j]/255.)*255.
            error = image[i,j] - cut_image
            if not flip:
                image[i, j + 1] += error*(7./16.)
                image[i + 1, j-1] += error*(3./16.)
                image[i + 1, j ] += error*(5./16.)
                image[i + 1, j + 1] += error*(1./16.)
            else:
                image[i, j - 1] += error*(7./16.)
                image[i + 1, j-1] += error*(1./16.)
                image[i + 1, j ] += error*(5./16.)
                image[i + 1, j + 1] += error*(3./16.)
    image = linear_rescale(image) 
    image = np.where(image > 128, 1, 0)
    return image
            
```

```python
# Run (this might take a while) ...
floyd_a = [floyd_steinberg_dithering(np.copy(images[index]),'a')\
           for index in range(len(image_names))]
floyd_b1 = [floyd_steinberg_dithering(np.copy(images[index]),'b1')\
           for index in range(len(image_names))]
floyd_b2 = [floyd_steinberg_dithering(np.copy(images[index]),'b2')\
           for index in range(len(image_names))]
```

## Save images

### Save the images in .pbm

```python
def save_pbm_images(images, files, save_dir):
    if not os.path.exists(save_dir):
        print ('Error the folder does not exist!')
        return
    for image, file in zip (images, files):
        cv2.imwrite(os.path.join(save_dir,file + '.pbm'), image, (cv2.IMWRITE_PXM_BINARY,0))
    print('Finished')
```

```python
base_dir = os.path.join('outputs','pbm_images')
save_dirs = ['ordered_dithering_a', 'ordered_dithering_b',
             'ordered_dithering_bayes_a','ordered_dithering_bayes_b',
             'floyd_steinberg_a', 'floyd_steinberg_b1','floyd_steinberg_b2'
            ]
save_images = [ordered_dithering_a,ordered_dithering_b,
               ordered_dithering_bayes_a, ordered_dithering_bayes_b,
               floyd_a, floyd_b1,floyd_b2
              ]
pbm_names = [name.split('.')[0] for name in image_names]

# Create the folders if they dont exist
for save_dir in save_dirs:
    if not os.path.exists(os.path.join(base_dir,save_dir)):
        print('Creating folder', os.path.join(base_dir,save_dir))
        os.makedirs(os.path.join(base_dir,save_dir))
```

```python
for save_dir, save_image_list in zip(save_dirs,save_images):
    save_pbm_images(save_image_list, pbm_names, os.path.join(base_dir,save_dir))
```

### Save the images in .png

```python
def save_png_images(images, files, save_dir):
    """ Save a list of images
    Arguments:
        images: a list of images
        files: a list of filenames
        save_dir: the folder name
    Return:
        None
    """
    if not os.path.exists(save_dir):
        print ('Error the folder does not exist!')
        return
    for image, file in zip (images, files):
        image = np.where(image == 1, 255, 0)
        cv2.imwrite(os.path.join(save_dir,file + '.png'), image)
    print('Finished') 
```

```python
base_dir = os.path.join('outputs','png_images')
save_dirs = ['ordered_dithering_a', 'ordered_dithering_b',
             'ordered_dithering_bayes_a','ordered_dithering_bayes_b',
             'floyd_steinberg_a', 'floyd_steinberg_b1','floyd_steinberg_b2'
            ]
save_images = [ordered_dithering_a,ordered_dithering_b,
               ordered_dithering_bayes_a, ordered_dithering_bayes_b,
               floyd_a, floyd_b1, floyd_b2
              ]
png_names = [name.split('.')[0] for name in image_names]

# Create the folders if they dont exist
for save_dir in save_dirs:
    if not os.path.exists(os.path.join(base_dir,save_dir)):
        print('Creating folder', os.path.join(base_dir,save_dir))
        os.makedirs(os.path.join(base_dir,save_dir))
```

```python
for save_dir, save_image_list in zip(save_dirs,save_images):
    save_png_images(save_image_list, png_names, os.path.join(base_dir,save_dir))
```
