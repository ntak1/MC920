## Part 1.1

In this part we are going to import the convolutional filters suggested in [here](http://www.ic.unicamp.br/~helio/disciplinas/MC920/trabalho1.pdf).

```python
# Defining the convolutional filters
h1, h2, h3, h4 = load_filters()

# Load Images
input_images, image_shape, input_images_names = load_images()

# Visualize the input images
color = 'gray'
titles = [name.split('.')[0] for name in input_images_names]
plot_images(input_images,titles=titles,num_columns=5,resize=15)

# Sanity check
print("Number of images: ", input_images.shape[0])
print("Shape of images: ", image_shape)
```

```python
# Convolve the image with the filters
filters = [h1, h2, h3, h4]

# Create a volume to store outputs
filtered_images = np.zeros((len(input_images_names),len(filters),*input_images[0].shape))

image_index = filter_index = 0
for image in input_images:
    filter_index = 0
    for conv_filter in filters:
        filtered_images[image_index,filter_index] = signal.convolve2d(image, conv_filter, mode='same')
        filter_index += 1
    image_index += 1
```

### Visualize the convolutions outputs

```python
# Defining a friendly notation
baboon, butterfly, city, house, seagull = range(5)
h1_index, h2_index, h3_index, h4_index = range(4)

# Choose a image to visualize
image = baboon

titles = ['h1', 'h2', 'h3','h4']
plot_images(filtered_images[baboon],resize=3,num_columns=2,titles=titles)
```

```python
# Chose a filter and visualize the results
for index in range(5):
    plot_image(linear_rescale(filtered_images[index,h4_index]),axis='off')
```

## Part 1.2
Adding h3 and h4 outputs
$$
    h_{34} = \sqrt{(h_{3})^{2}+ (h_{4})^{2}}
$$

```python
# Add the filters response according to the above equation
weighted_volume= np.zeros((len(input_images_names),*image_shape))
for image_index in range(weighted_volume.shape[0]):
     weighted_volume[image_index,:,:] = np.sqrt((filtered_images[image_index, h3_index]**2)\
                + (filtered_images[image_index, h4_index]**2))
```

```python
titles = ['h3_h4_' + name.split('.')[0] for name in input_images_names]
print(titles)
```

```python
# Visualize the output
weighted_volume = [linear_rescale(img).astype('int32') for img in weighted_volume]
titles = ['h3_h4_' + name.split('.')[0] for name in input_images_names]
i = 0
for img in np.array(weighted_volume):
    plot_image(linear_rescale(img), axis='off',
#                savename = os.path.join('output_imgs', 'part1',titles[i] + '.png')
              )
    print(os.path.join('output_imgs', 'part1',titles[i]))
    i+=1
```

# Part 2: Fourier
Using Scipy's Fourier filter [here](https://docs.scipy.org/doc/scipy-0.16.1/reference/generated/scipy.ndimage.fourier.fourier_gaussian.html).

```python
# Apply the Gaussian filter in frequency domain using a scipy function
sigma = 1
fourier_images = [fourier.fourier_gaussian(image,sigma=sigma) for image in input_images]
fourier_images = np.array(fourier_images)
```

```python
# Visualize the same sigma to different images
titles = ['baboon', 'butterfly', 'city','house', 'seagull']
titles = [title + ', sigma = ' + str(sigma) for title in titles]
plot_images(fourier_images,num_columns=2, resize=3, titles=titles)
```

```python
# Apply different sigma values to an image
sigmas = [0.1, 0.5, 2, 10]
sigmas_outputs = [fourier.fourier_gaussian(input_images[1],sigma=sigma) for sigma in sigmas]
i = 0
for img in sigmas_outputs:
    plot_image(img,axis='off')
    i+=1
```

### 3.1 Board detection

```python
# Generate a black and white pattern
white = 255
black = 0

stripe_color = white
background_color = black
cross_image = np.full((image_shape),background_color)
num_stripes = 1
breadth = 51

stripe_indices = np.arange(0,image_shape[0],image_shape[0]//(num_stripes+1))
stripe_indices = stripe_indices[stripe_indices>0]
for i in range(-breadth//2, breadth//2+1):
    cross_image[stripe_indices+ i,:] = stripe_color
    cross_image[:,stripe_indices+ i] = stripe_color

plot_image(cross_image)
```

```python
conv_filter = h4

output1 =  linear_rescale(signal.convolve2d(cross_image,conv_filter,mode='same'))
output2 = np.clip(np.abs(signal.convolve2d(cross_image,conv_filter,mode='same')),0,255)

plot_image(output2,savename='h4_clip_extra.png')
```
